## Lister
Du kan skrive punktlister ved å bruke `*` eller `-` foran hvert element. For eksempel blir
```
* Dette er et punkt
* Dette er et nytt punkt
```
formatert som
* Dette er et punkt
* Dette er et nytt punkt

Du kan også lage numererte lister. For eksempel blir
```
1. Parenteser
2. Potenser
3. Gange og dele
4. Pluss og minus
```
formatert som
1. Parenteser
2. Potenser
3. Gange og dele
4. Pluss og minus

Du kan kombinere lister ved å bruke innrykk. For eksempel blir (legg merke til at du kan bruke hvilke som helst tall)
```
1. Noen mattefolk
    1. Gauss
    1. Euler
    1. Euklid
1. Noen informatikkfolk
    1. Turing
    1. Dijkstra
    1. Knuth
```
formatert som
1. Noen mattefolk
    1. Gauss
    1. Euler
    1. Euklid
1. Noen informatikkfolk
    1. Turing
    1. Dijkstra
    1. Knuth
    
## Formatere tekst
For å skrive tekst i _kursiv_, setter du en enkelt understrek på hver side av det du vil formatere. For eksempel er det _svært viktig_ å huske at $\lim_{v\to0}\frac{\sin v}{v} = 1$ kun fungerer om $v$ er i _radianer_. (Her har jeg for eksempel skrevet `_svært viktig_`.

For å __utheve__ skrift, gjør du det samme, men nå med dobbel understrek.

__Obs!__ Vi definerer tangensfunksjonen som $\tan v = \frac{\sin v}{\cos v}$. (Her har jeg skrevet `__Obs!__`.)

## Formatere matte
Om du ønsker å skrive matte sammen med den øvrige teksten, kan du gjøre dette ved å bruke enkle dollartegn, $\pi = 3.1415\ldots$ (som er skrevet $\pi = 3.1415\ldots$). Om du ønsker å formatere matematikken på egne linjer, som du bør om du skriver likninger, for eksempel, kan du gjøre dette med dobbelte dollartegn. For eksempel blir `$$x_{1,2} = \frac{-b\pm\sqrt{b^2 - 4ac}}{2a}$$` formatert som $$x_{1,2} = \frac{-b\pm\sqrt{b^2 - 4ac}}{2a}$$

[Denne lenka](https://docx2latex.com/tutorials/Mathematical-equations-LaTeX.html) viser en delvis oversikt over hva du kan gjøre med slik matteskriving. Dette er basert på et formateringsspråk som heter $\LaTeX$ (Latex), og er en slags standard for hvordan man skriver matematikk på en datamaskin. Du kan også bruke et [onlineverktøy](http://www.sciweavers.org/free-online-latex-equation-editor) for å lage likninger og uttrykk.

## Sitater og definisjoner

```

```

Blir det formatert som

>
